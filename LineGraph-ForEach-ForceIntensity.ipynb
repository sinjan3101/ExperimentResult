```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from Excel
df = pd.read_excel('TaskThresholdComparison.xlsx')

# Rename columns for easier access
df.columns = [
    'Force Intensity',
    'Task1',
    'Task2',
    'Task3'
]

# Mapping of full task names
full_task_names = [
    'Task1: Pushing at a fixed point',
    'Task2: Pushing and rolling over the surface',
    'Task3: Pushing along the edges'
]

# Plotting
conditions = ['Task1', 'Task2', 'Task3']
labels = df['Force Intensity']
colors = ['chocolate', 'magenta', 'red', 'blue']
#colors = ['#3db6e3', '#f7c59f', '#c7e9b0', '#6a9fb5']

plt.figure(figsize=(14, 8))
for i, row in df.iterrows():
    plt.plot(conditions, row[1:], marker='o', label=labels[i], color=colors[i])

#plt.xlabel('Task', fontsize=22)
plt.ylabel('Refresh rate (Hz)', fontsize=22)
#plt.title('Refresh rate by force', fontsize=22)
plt.legend(title='Force Intensity', fontsize=21, title_fontsize=22, loc='lower right')
plt.grid(True, alpha=0.3)
plt.xticks(fontsize=22)
plt.yticks(fontsize=22)

# List full task names on the right side of the plot
for i, name in enumerate(full_task_names):
    plt.text(0, 900 - (i)*50, name, fontsize=22, va='bottom')
    #plt.text(0, 400 - (i+3)*100, name, fontsize=22, va='bottom')

plt.tight_layout()
plt.savefig('Minimum_Refresh_Rate_Foreach_ForceIntensity_LineGraph.png')
plt.show()
```
