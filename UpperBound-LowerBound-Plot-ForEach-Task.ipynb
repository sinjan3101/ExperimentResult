```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from both sheets (assuming the file name based on attachment)
file_path = 'TaskThresholdComparison_upper_bound.xlsx'  # Adjust if needed

lower_bound_df = pd.read_excel(file_path, sheet_name='lower-bound')
upper_bound_df = pd.read_excel(file_path, sheet_name='upper-bound')

# Rename columns for clarity (consistent for both sheets)
columns = [
    'Force Intensity',
    'Pushing at a fixed point',
    'Pushing alongwith rolling over the surface',
    'Pushing along the edges'
]
lower_bound_df.columns = columns
upper_bound_df.columns = columns

# Define plotting parameters
force_levels = lower_bound_df['Force Intensity']
tasks = [
    #'Pushing at a fixed point'
    'Pushing alongwith rolling over the surface'
    #'Pushing along the edges'
]

# Colors: base colors for tasks, with variations for bounds
#base_colors = ['#3db6e3', '#f7c59f', '#c7e9b0']  # Lower bound
#upper_colors = ['#1a8bc4', '#d89f7a', '#a3c98a']  # Upper bound (darker shades for distinction)
base_colors = ['Red', '#f7c59f', '#c7e9b0']
upper_colors = ['Blue', '#d89f7a', '#a3c98a']
# Create the plot
plt.figure(figsize=(14, 5))

for i, task in enumerate(tasks):
    # Plot lower bound with solid line
    plt.plot(force_levels, lower_bound_df[task], marker='x', linestyle='--', 
             label=f'Lower Bound: {task}', color=base_colors[i])
    
    # Plot upper bound with dashed line
    plt.plot(force_levels, upper_bound_df[task], marker='o', linestyle='-', 
             label=f'Upper Bound: {task}', color=upper_colors[i])

plt.xlabel('Force Intensity', fontsize=22)
plt.ylabel('Refresh Rate (Hz)', fontsize=22)
plt.grid(True, alpha=0.5)
plt.xticks(fontsize=22, rotation=15)
plt.yticks(range(0, 1100, 100), fontsize=22)
plt.legend(title='Tasks and Bounds', fontsize=14, title_fontsize=19, loc='lower right')
plt.tight_layout()
plt.savefig('RefreshRateComparison.png')  # Saves the figure
plt.show()
```
